# Binary Search Tree data structure

```python
class Node:
    def __init__(self,data=None,left=None,right=None):
        self.data = data
        self.left=left
        self.right=right

class BST:
    def __init__(self):
        self.root=None

    # insert data using recursion
    def insert(self,data):
        self.root=self.rinsert(self.root,data)

    def rinsert(self,root,data):
        if root is None:
            return Node(data)
        if data<root.data:
            root.left=self.rinsert(root.left,data)
        elif data>root.data:
            root.right=self.rinsert(root.right,data)
        return root

    # insert data using iteration
    def insert_data(self,data):
        if self.root is None:
            self.root=Node(data=data)
        else:
            temp=self.root
            while True:
                if temp.data<data:
                    if temp.right is None:
                        new_node=Node(data=data)
                        temp.right=new_node
                        break
                    else:
                        temp=temp.right
                else:
                    if temp.left is None:
                        new_node=Node(data=data)
                        temp.left=new_node
                        break
                    else:
                        temp=temp.left

    def search(self,data):
        return self.rsearch(self.root,data)
        
    def rsearch(self,root,data):
        if root is None:
            return 
        if root is not None and root.data==data:
            return True
        elif data<root.data:
            return self.rsearch(root.left,data)
        else:
            return self.rsearch(root.right,data)
    
    def inorder_traversal(self):
        result=[]
        self.rinorder(self.root,result)
        return result
    
    def rinorder(self,root,result):
        if root:
            self.rinorder(root.left,result)
            result.append(root.data)
            self.rinorder(root.right,result)
        
    def preorder_traversal(self):
        result=[]
        self.rpreorder(self.root,result)
        return result
    
    def rpreorder(self,root,result):
        if root:
            result.append(root.data)
            self.rpreorder(root.left,result)
            self.rpreorder(root.right,result)
        
    def postorder_traversal(self):
        result=[]
        self.rpostorder(self.root,result)
        return result
    
    def rpostorder(self,root,result):
        if root:
            self.rpostorder(root.left,result)
            self.rpostorder(root.right,result)
            result.append(root.data)

    def min_value(self):
        current=self.root
        while current.left is not None:
            current=current.left
        return current.data

    def max_value(self):
        current=self.root
        while current.right is not None:
            current=current.right
        return current.data 
    
    def delete(self,data):
        self.root=self.rdelete(self.root,data)

    def rdelete(self,root,data):
        if root is None:
            return root
        if data<root.data:
            root.left=self.rdelete(root.left,data=data)
        elif data>root.data:
            root.right=self.rdelete(root.right,data=data)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            
            # replace the value with sucessor
            root.data=self.min_value(root.right)
            root.right=self.rdelete(root.right,root.data)

        return root

    def size(self):
        return len(self.inorder_traversal())

        
    # def traversing(self):
    #     child=self.root
    #     parent=self.root
    #     while True:
    #         if child.left is None:
    #             print(child.data)
    #             if child.right is None:
    #                 while parent.left!=child or parent.right!=child:
    #                     pass

t1=BST()
t1.insert_data(10)
t1.insert_data(5)
t1.insert_data(12)
t1.insert_data(2)
t1.insert_data(7)
t1.insert_data(11)
t1.insert_data(14)
# t1.insert(40)
print(t1.inorder_traversal())
print(t1.preorder_traversal())
print(t1.postorder_traversal())
print(t1.search(9))
print(t1.min_value())
print(t1.max_value())
t1.delete(7)
print(t1.inorder_traversal())
print(t1.size())
```
