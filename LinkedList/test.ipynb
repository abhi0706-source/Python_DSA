# Singly Linked List

```python
# Define Singly linked list with full operations

class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class SLL:
    def __init__(self,start=None):
        self.start=start 

    def is_empty(self):
        if self.start==None:
            return True
        return False
        
    def insert_at_start(self,data):
        new_node=Node(data=data,next=self.start)
        self.start=new_node

    def insert_at_last(self,data):
        new_node=Node(data=data)
        if not self.is_empty():
            temp=self.start
            while temp.next!=None:
                temp=temp.next
            temp.next=new_node
        else:
            self.start=new_node
    
    def search(self,data):
        temp=self.start
        while temp is not None:
            if temp.data==data:
                return temp
            temp=temp.next
        return None
    
    def insert_after(self,temp,data):
        if temp is not None:
            new_node=Node(data=data,next=temp.next)
            temp.next=new_node

    def print_list(self):
        temp=self.start
        while temp is not None:
            print(temp.data,end=" -> ")
            temp=temp.next

    def delete_first(self):
        if self.start is not None:
            self.start=self.start.next

    def delete_last(self):
        if self.start is None:
            pass
        elif self.start.next is None:
            self.start=None
        else:
            temp=self.start
            while temp.next is not None:
                if temp.next.next==None:
                    break
                temp=temp.next
            
            temp.next=None

    def delete_item(self,data):
        
        if self.start is None:
            pass
        elif self.start.next is None:
            if self.start.data==data:
                self.start=None
        else:
            temp=self.start
            if self.start.data==data:
                self.start=self.start.next
            else:
                while temp.next is not None:
                    if temp.next.data==data:
                        temp.next=temp.next.next
                        break
                    temp=temp.next 
    def __iter__(self):
        return SLLIterator(self.start)

class SLLIterator:
    def __init__(self,start):
        self.current=start

    def __iter__(self):
        return self
    
    def __next__(self):
        if not self.current:
            raise StopIteration
        data=self.current.data
        self.current=self.current.next
        return data


        

mylist=SLL()
mylist.insert_at_start(20)
mylist.insert_at_start(10)
mylist.insert_at_last(30)
mylist.insert_after(mylist.search(20),25)
mylist.print_list()
print()
print("*******************************")
# mylist.delete_item(20)
# mylist.delete_last()
for i in mylist:
    print(i,end="=>")
# mylist.print_list()
```

# Circular Linked-list

```python
# Define Circular-Singly linked list with full operations

class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class CLL:
    def __init__(self,last=None):
        self.last=last

    def is_empty(self):
        if self.last==None:
            return True
        return False
        
    def insert_at_start(self,data):
        if self.is_empty():
            new_node=Node(data=data)
            self.last=new_node
            new_node.next=self.last
        else:
            new_node=Node(data=data,next=self.last.next)
            self.last.next=new_node

    def insert_at_last(self,data):
        new_node=Node(data=data,next=self.last.next)
        if not self.is_empty():
            temp=self.last.next
            while temp.next!=self.last.next:
                temp=temp.next
            temp.next=new_node
            self.last=new_node
        else:
            self.last=new_node
    
    def search(self,data):
        temp=self.last.next
        print("Inside search",temp.data)
        while temp is not self.last:
            if temp.data==data:
                return temp
            temp=temp.next
        return None
    
    def insert_after(self,temp,data):

        print("Temp data:",temp.data)

        if temp is not self.last.next:
            new_node=Node(data=data,next=temp.next)
            temp.next=new_node

    def print_list(self):
        temp=self.last.next
        while temp is not self.last:
            print(temp.data,end=" -> ")
            temp=temp.next
        print(temp.data,end=" -> ")

    def delete_first(self):
        if self.last.next is not None:
            self.last.next=self.last.next.next

    def delete_last(self):
        if self.last is None:
            pass
        elif self.last is self.last.next:
            self.last=None
        else:
            temp=self.last.next
            while temp.next is not self.last:
                print("Inside loop",temp.data)
                temp=temp.next
            print("outside loop",temp.data)
            
            temp.next=self.last.next
            self.last=temp

    def delete_item(self,data):
        
        if self.last is None:
            pass
        elif self.last is self.last.next:
            if self.last.data==data:
                self.last=None
        else:
            temp=self.last.next
            if self.last.next.data==data:
                self.last.next=self.last.next.next
            else:
                while temp.next is not self.last.next:
                    if temp.next.data==data:
                        temp.next=temp.next.next
                        break
                    temp=temp.next 
    def __iter__(self):
        return CLLIterator(self.last.next)

class CLLIterator:
    def __init__(self,start):
        self.current=start

    def __iter__(self):
        return self
    
    def __next__(self):
        if not self.current:
            raise StopIteration
        data=self.current.data
        self.current=self.current.next
        return data


print("Hello")
mylist=CLL()
print("********************************")
mylist.insert_at_start(20)
mylist.insert_at_start(10)
mylist.insert_at_last(50)
mylist.insert_after(mylist.search(20),25)
# mylist.print_list()
mylist.print_list()
print()
print("*******************************")
mylist.delete_item(20)
mylist.delete_last()
mylist.print_list()
mylist.delete_first()
print()
mylist.print_list()
# for i in mylist:
#     print(i,end="=>")
# mylist.print_list()
```

```python
class Node:
    def __init__(self,data=None,prev=None,next=None):
        self.data=data
        self.next=next
        self.prev=prev

class CDLL:
    def __init__(self,start=None) -> None:
        self.start=start


    def insterstion_at_first(self,data):
        if self.start is not None: 
            new_node=Node(data=data,next=self.start,prev=self.start.prev)
            self.start.prev.next=new_node
            self.start.next.prev=new_node
            self.start=new_node
        else:
            new_node=Node(data=data)
            self.start=new_node
            self.start.prev=self.start
            self.start.next=self.start

    def insterstion_at_last(self,data):
        if self.start is not None: 
            temp=self.start
    
            while temp is not self.start.prev:
                temp=temp.next
            new_node=Node(data=data,next=temp.next,prev=temp)
            temp.next.prev=new_node
            temp.next=new_node

        else:
            new_node=Node(data=data)
            self.start=new_node
            self.start.prev=self.start
            self.start.next=self.start
    
    def insert_after(self,data,index_val):
        if self.start is not None:
            temp=self.start
            while temp.next is not self.start.prev:
                if temp.data==index_val:
                    # print(temp.data)
                    new_node=Node(data=data,prev=temp,next=temp.next)
                    temp.next.prev=new_node
                    temp.next=new_node
                    break
                temp=temp.next

    def search_data(self,data):
        temp=self.start
        index=1
        while temp is not self.start.prev:
            if temp.data==data:
                return index
            temp=temp.next
            index+=1

            if temp.data==data:
                return index
            # print(temp.data)
        return -1

    def print_list(self):
        temp=self.start
        while temp is not self.start.prev:
            print(temp.data,end=" ")
            temp=temp.next
        print(temp.data,end=" ")

    def is_empty(self):
        return self.start is None
    
    def delete_at_first(self):
        if self.start is None:
            pass
        elif self.start.next is self.start:
            self.start=None
        else:
            self.start.prev.next=self.start.next
            self.start.next.prev=self.start.prev
            self.start=self.start.next

    def delete_at_last(self):
        if self.start is None:
            pass
        elif self.start.next is self.start:
            self.start=None
        else:
            temp=self.start
            while temp is not self.start.prev:
                temp=temp.next

            temp.next.prev=temp.prev
            temp.prev.next=temp.next

    def delete_specific_item(self,data):
        if self.start is None:
            pass
        elif self.start.next is self.start:
            if self.start.data==data:
                self.start=None
        else:
            temp=self.start
            while temp.next is not self.start.prev:
                if temp.data==data:
                    temp.prev.next=temp.next
                    temp.next.prev=temp.prev
                    break
                temp=temp.next

    def __iter__(self):
        return CDLLIterator(self.start)



class CDLLIterator:
    def __init__(self, start):
        self.start=start
        self.current=start
        self.count=0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current is None:
            raise StopIteration
        
        if self.current==self.start and self.count==1:
            raise StopIteration
        else:
            self.count=1

        data=self.current.data
        self.current=self.current.next
        return data



my_dll_list=CDLL()
# my_dll_list.insterstion_at_first(data=10)
# my_dll_list.insterstion_at_first(data=20)
# my_dll_list.print_list()
print(my_dll_list.is_empty())
my_dll_list.insterstion_at_last(data=40)
my_dll_list.insterstion_at_last(data=30)
my_dll_list.insterstion_at_last(data=25)
my_dll_list.insterstion_at_last(data=50)
my_dll_list.insterstion_at_last(data=10)
my_dll_list.print_list()
print()
print("---------------------------------------------")
my_dll_list.insterstion_at_first(data=20)
# my_dll_list.print_list()
print()
my_dll_list.insert_after(data=35,index_val=30)
my_dll_list.print_list()
print()
print(my_dll_list.is_empty())

print(my_dll_list.search_data(30))

for i in my_dll_list:
    print(i)

print("__________________________________________")
my_dll_list.print_list()
print()
my_dll_list.delete_at_first()
my_dll_list.print_list()
print()
print()
my_dll_list.delete_at_last()
my_dll_list.print_list()
print()
my_dll_list.delete_specific_item(35)
print()
my_dll_list.print_list()
print()
print("__________________________________________")
```
