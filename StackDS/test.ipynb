# Stack Data-Structure

## Stack data-structure using list

```python
class Stack:
    def __init__(self):
        self.stack=[]
    
    def is_empty(self):
        if len(self.stack)==0:
            return True
        return False

    def push(self,item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return "Stack is empty"
        return self.stack.pop()
    
    def peek(self):
        if self.is_empty():
            return "Stack is empty"
        return self.stack[-1]
    
    def size(self):
        return len(self.stack)


s1=Stack()
print(s1.pop()) 
print("Is empty",s1.is_empty()) # True
s1.push(1)
s1.push(2)
s1.push(3)
print(s1.peek()) 
print("Is empty",s1.is_empty()) 
print("pop element is:",s1.pop())
print(s1.peek())
s1.push(10)
s1.push(20)
print(s1.peek())
print(s1.size())
```

## Stack Data-structure by extending list class

```python
class Stack(list):
    def is_empty(self):
        if len(self)==0:
            return True
        return False

    def push(self,item):
        self.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return super().pop()
    
    def peek(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self[-1]
    
    def size(self):
        return len(self)
    
    def insert(self,index,data):
        raise AttributeError("No attribute insert in Stack.")


s1=Stack()
# print(s1.insert(3,5))
# print(s1.pop()) 
print("Is empty",s1.is_empty()) # True
s1.push(1)
s1.push(2)
s1.push(3)
print(s1.peek()) 
print("Is empty",s1.is_empty()) 
print("pop element is:",s1.pop())
print(s1.peek())
s1.push(10)
s1.push(20)
print(s1.peek())
print(s1.size())
```

## Implementing Stack using linked list concept

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class Stack:
    def __init__(self,top_val=None):
        self.top_val=top_val
        self.count=0

    def is_empty(self):
        if self.top_val==None:
            return True
        return False

    def push(self,data):
        if not self.is_empty():
            new_node=Node(data=data,next=self.top_val.next)
            temp=self.top_val.next
            while temp.next!=self.top_val.next:
                temp=temp.next
            temp.next=new_node
            self.top_val=new_node
        else:
            new_node=Node(data=data)
            self.top_val=new_node
            new_node.next=self.top_val
        
        self.count+=1
    

    def pop(self):
        if self.top_val is None:
            pass
        elif self.top_val is self.top_val.next:
            self.top_val=None
        else:
            temp=self.top_val.next
            while temp.next is not self.top_val:
                temp=temp.next
            
            pop_val=temp.next.data

            temp.next=self.top_val.next
            self.top_val=temp

        self.count-=1

        return pop_val


    def peek(self):
        return self.top_val.data
    
    def size(self):
        return self.count

s1=Stack() 
print("Is empty",s1.is_empty()) # True
s1.push(1)
s1.push(2)
s1.push(3)
print(s1.peek()) 
print("Is empty",s1.is_empty()) 
print("pop element is:",s1.pop())
print(s1.peek())
s1.push(10)
s1.push(20)
print(s1.peek())
print(s1.size())
```

## Stack using linked list using implemented class

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class SLL:
    def __init__(self,start=None):
        self.start=start
        self.count=0

    def is_empty(self):
        if self.start==None:
            return True
        return False

    def insert_at_start(self,data):
        if self.start is not None:
            new_node=Node(data=data,next=self.start)
            self.start=new_node
        else:
            new_node=Node(data=data)
            self.start=new_node
        self.count+=1
    
    def delete_first(self):
        if self.start is not None:
            pop_val=self.start.data
            self.start=self.start.next
            self.count-=1
            return pop_val

    def peek_list(self):
        return self.start.data
    
    def size_list(self):
        return self.count


class Stack:
    def __init__(self):
        self.stack=SLL()
    
    def is_empty(self):
        return self.stack.is_empty()

    def push(self,data):
        self.stack.insert_at_start(data=data)
    
    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.stack.delete_first()

    def peek(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.stack.peek_list()
    
    def size(self):
        return self.stack.size_list()


s1=Stack() 
print("Is empty",s1.is_empty()) # True
s1.push(1)
s1.push(2)
s1.push(3)
print(s1.peek()) 
print("Is empty",s1.is_empty()) 
print("pop element is:",s1.pop())
print(s1.peek())
s1.push(10)
s1.push(20)
print(s1.peek())
print(s1.size())
```

## Implement Stack by inherited linkedlist class

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class SLL:
    def __init__(self,start=None):
        self.start=start
        self.count=0

    def is_empty(self):
        if self.start==None:
            return True
        return False

    def insert_at_start(self,data):
        if self.start is not None:
            new_node=Node(data=data,next=self.start)
            self.start=new_node
        else:
            new_node=Node(data=data)
            self.start=new_node
        self.count+=1
    
    def delete_first(self):
        if self.start is not None:
            pop_val=self.start.data
            self.start=self.start.next
            self.count-=1
            return pop_val

    def peek_list(self):
        return self.start.data
    
    def size_list(self):
        return self.count


class Stack(SLL):
    
    def is_empty():
        return super().is_empty()

    def push(self,data):
        self.insert_at_start(data=data)
    
    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.delete_first()

    def peek(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.peek_list()
    
    def size(self):
        return self.size_list()
    
s1=Stack() 
print("Is empty",s1.is_empty()) # True
s1.push(1)
s1.push(2)
s1.push(3)
print(s1.peek()) 
print("Is empty",s1.is_empty()) 
print("pop element is:",s1.pop())
print(s1.peek())
s1.push(10)
s1.push(20)
print(s1.peek())
print(s1.size())
```
