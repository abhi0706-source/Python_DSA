## Deque using list way

```python
class Deque:
    def __init__(self) -> None:
        self.deque = []

    def is_empty(self)->None:
        return len(self.deque) == 0
    
    def insert_front(self,data=None)->None:
        self.deque.insert(0, data)

    def insert_rear(self,data=None)->None:
        self.deque.append(data)

    def delete_front(self)->None:
        if self.is_empty():
            raise IndexError("Deque is empty")
        else:
            self.deque.pop(0)

    def delete_rear(self)->None:
        if self.is_empty():
            raise IndexError("Deque is empty")
        else:
            self.deque.pop()

    def get_front(self)->None:
        if self.is_empty():
            raise IndexError("Deque is empty")
        else:
            return self.deque[0]
    
    def get_rear(self)->None:
        if self.is_empty():
            raise IndexError("Deque is empty")
        else:
            return self.deque[len(self.deque)-1]
        
    def size(self)->None:
        return len(self.deque)
    

d1=Deque()
d1.insert_front(20)
d1.insert_front(30)
d1.insert_front(40)
d1.insert_rear(10)
d1.insert_rear(50)
print(d1.get_front())
print(d1.get_rear())
d1.delete_front()        
d1.delete_rear()        
print(d1.get_front())
print(d1.get_rear())
```

## Apply Deque using doubly linkedlist

```python
class Node:
    def __init__(self,data=None,prev=None,next=None):
        self.data=data
        self.next=next
        self.prev=prev

class Deque:
    def __init__(self):
        self.front=None
        self.rear=None
        self.count=0
 
    def is_empty(self):
        return self.count==0
    
    def insert_front(self,data):
        new_node=Node(data=data,next=self.front)
        if self.front is None:
            self.rear=new_node
            self.front=new_node
        else:
            self.front.prev=new_node
            self.front=new_node

        self.count+=1

    def insert_rear(self,data):
        new_node=Node(data=data,prev=self.rear)
        if self.rear is None:
            self.front=new_node
        else:
            self.rear.next=new_node
            self.rear=new_node
        
        self.count+=1

    def delete_front(self):
        if self.is_empty():
            raise IndexError("Deque is empty")
        if self.rear==self.front:
            self.front=None
            self.rear=None
        else:
            self.front=self.front.next
            self.front.prev=None
        
        self.count-=1

    def delete_rear(self):
        if self.is_empty():
            raise IndexError("Deque is empty")
        if self.rear==self.front:
            self.front=None
            self.rear=None
        else:
            self.rear=self.rear.prev
            self.rear.next=None
        
        self.count-=1

    def get_front(self)->None:
        if self.is_empty():
            raise IndexError("Deque is empty")
        else:
            return self.front.data
    
    def get_rear(self)->None:
        if self.is_empty():
            raise IndexError("Deque is empty")
        else:
            return self.rear.data
        
    def size(self)->None:
        return self.count
    



d1=Deque()
d1.insert_front(20)
d1.insert_front(30)
d1.insert_front(40)
d1.insert_rear(10)
d1.insert_rear(50)
d1.insert_rear(70)
print(d1.get_front())
print(d1.get_rear())
d1.delete_front()        
d1.delete_rear()        
print(d1.get_front())
print(d1.get_rear())
```
