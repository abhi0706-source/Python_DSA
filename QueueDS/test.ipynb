# Queue Data-Structure

## Queue Data-Struecture using list

```python
class Queue:
    def __init__(self):
        self.queue=[]
    
    def is_empty(self):
        if len(self.queue)==0:
            return True
        return False

    def enqueue(self,item):
        self.queue.insert(0,item)

    def dequeue(self):
        if self.is_empty():
            return "queue is empty"
        return self.queue.pop()
    
    def get_front(self):
        if self.is_empty():
            return "queue is empty"
        return self.queue[0]
    
    def get_rear(self):
        if self.is_empty():
            return "queue is empty"
        return self.queue[-1]
    
    def size(self):
        return len(self.queue)


q1=Queue()
q1.enqueue(20)
q1.enqueue(30)
q1.enqueue(40)
print("Get Front",q1.get_front())
print("Get Rear",q1.get_rear())
print("dequeue element:",q1.dequeue())
print("Get Rear",q1.get_rear())
print("Get Size of queue",q1.size())
```

## Queue Data-structure by extending list class

```python
class Queue(list):
    def is_empty(self):
        if len(self)==0:
            return True
        return False

    def enqueue(self,item):
        self.insert(1,item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return super().pop()
    
    def get_front(self):
        if self.is_empty():
            return "queue is empty"
        return self[0]
    
    def get_rear(self):
        if self.is_empty():
            return "queue is empty"
        return self[-1]
    
    def size(self):
        return len(self)



q1=Queue()
q1.enqueue(20)
q1.enqueue(30)
q1.enqueue(40)
print("Get Front",q1.get_front())
print("Get Rear",q1.get_rear())
print("dequeue element:",q1.dequeue())
print("Get Rear",q1.get_rear())
print("Get Size of queue",q1.size())
```

## Implementing Queue using linked list concept

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class Queue:
    def __init__(self,front=None,rear=None):
        self.front=front
        self.rear=rear
        self.count=0

    def is_empty(self):
        if self.front==None and self.rear==None:
            return True
        return False

    def enqueue(self,data):
        if not self.is_empty():
            new_node=Node(data=data)
            self.rear.next=new_node
            self.rear=new_node
        else:
            new_node=Node(data=data)
            self.front=new_node
            self.rear=new_node
        self.count+=1
    

    def dequeue(self):
        if self.front is None:
            pass
        elif self.front==self.rear :
            self.front=None
            self.rear=None
        else:
            dequeue_val=self.front.data
            self.front=self.front.next
        self.count-=1

        return dequeue_val
   
    def get_front(self):
        if self.is_empty():
            return "queue is empty"
        return self.front.data
    
    def get_rear(self):
        if self.is_empty():
            return "queue is empty"
        return self.rear.data
    
    def size(self):
        return self.count

q1=Queue()
q1.enqueue(20)
q1.enqueue(30)
q1.enqueue(40)
print("Get Front",q1.get_front())
print("Get Rear",q1.get_rear())
print("dequeue element:",q1.dequeue())
print("Get Rear",q1.get_rear())
print("Get Size of queue",q1.size())
```

## Queue using linked list using implemented class

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class SLL:
    def __init__(self,start=None):
        self.start=start
        self.count=0

    def is_empty(self):
        if self.start==None:
            return True
        return False

    def insert_at_last(self,data):
        new_node=Node(data=data)
        if not self.is_empty():
            temp=self.start
            while temp.next!=None:
                temp=temp.next
            temp.next=new_node
        else:
            self.start=new_node
    
        self.count+=1
    
    def delete_first(self):
        if self.start is not None:
            dequeue_val=self.start.data
            self.start=self.start.next
            self.count-=1
            return dequeue_val
    
    def get_front_val(self):
        if self.is_empty():
            return "queue is empty"
        return self.start.data
    
    def get_rear_val(self):
        if self.is_empty():
            return "queue is empty"
        temp=self.start
        while temp.next is not None:
            temp=temp.next
        return temp.data
    
    def size_queue(self):
        return self.count


class Queue:
    def __init__(self):
        self.queue=SLL()
    
    def is_empty(self):
        return self.queue.is_empty()

    def enqueue(self,data):
        self.queue.insert_at_last(data=data)
    
    def dequeue(self):
        if self.is_empty():
            raise IndexError("queue is empty")
        return self.queue.delete_first()

    def get_front(self):
        if self.is_empty():
            raise IndexError("queue is empty")
        return self.queue.get_front_val()
    
    def get_rear(self):
        if self.is_empty():
            raise IndexError("queue is empty")
        return self.queue.get_rear_val()
    
    def size(self):
        return self.queue.size_queue()


q1=Queue()
q1.enqueue(20)
q1.enqueue(30)
q1.enqueue(40)
print("Get Front",q1.get_front())
print("Get Rear",q1.get_rear())
print("dequeue element:",q1.dequeue())
print("Get Rear",q1.get_rear())
print("Get Size of queue",q1.size())
```

## Implement Queue by inherited linkedlist class

```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class SLL:
    def __init__(self,start=None):
        self.start=start
        self.count=0

    def is_empty(self):
        if self.start==None:
            return True
        return False

    def insert_at_last(self,data):
        new_node=Node(data=data)
        if not self.is_empty():
            temp=self.start
            while temp.next!=None:
                temp=temp.next
            temp.next=new_node
        else:
            self.start=new_node
    
        self.count+=1
    
    def delete_first(self):
        if self.start is not None:
            dequeue_val=self.start.data
            self.start=self.start.next
            self.count-=1
            return dequeue_val
    
    def get_front_val(self):
        if self.is_empty():
            return "queue is empty"
        return self.start.data
    
    def get_rear_val(self):
        if self.is_empty():
            return "queue is empty"
        temp=self.start
        while temp.next is not None:
            temp=temp.next
        return temp.data
    
    def size_queue(self):
        return self.count

class Stack(SLL):
    
    def is_empty(self):
        return super().is_empty()

    def push(self,data):
        self.insert_at_last(data=data)
    
    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.delete_first()
    
    def get_front(self):
        if self.is_empty():
            raise IndexError("queue is empty")
        return self.get_front_val()
    
    def get_rear(self):
        if self.is_empty():
            raise IndexError("queue is empty")
        return self.get_rear_val()
    
    def size(self):
        return self.size_queue()
    
q1=Queue()
q1.enqueue(20)
q1.enqueue(30)
q1.enqueue(40)
print("Get Front",q1.get_front())
print("Get Rear",q1.get_rear())
print("dequeue element:",q1.dequeue())
print("Get Rear",q1.get_rear())
print("Get Size of queue",q1.size())
```
